### The temperture given here is in fahrenheit.Convert it into Celsius using the formula Celsius=(Fahrenheit-32)* (5/9)

```python
Fahrenheit=GYM['temperature']
```

### Converting it into the list so we can apply lambda function

```python
F=Fahrenheit.tolist()
```

### Applying Lambda function

```python
C= map(lambda x: (float(5)/9)*(x-32),F)
Celsius=(list(C))
```

### Converting list to series

```python
temperature_celsius=pd.Series(Celsius)
```

### Applying the series to temperature column

```python
GYM['temperature']= temperature_celsius
GYM['temperature']
GYM.head()
```

```python
X = GYM.iloc[:,3:8]  # all rows, all the features and no labels
y = GYM.iloc[:, 0]  # all rows, label only
#X
#y
```

### Standardising data

```python
# Scale the data to be between -1 and 1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X= scaler.fit_transform(X)
X
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit_transform(X)
```

```python
pca.get_covariance()
```

```python
pca.get_covariance()
```

```python
explained_variance=pca.explained_variance_ratio_
explained_variance
```

```python
with plt.style.context('dark_background'):
    plt.figure(figsize=(8, 4))

    plt.bar(range(5), explained_variance, alpha=0.5, align='center',
          label='individual explained variance')
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc='best')
    plt.tight_layout()
```

### Thus we can see from the above plot that first two components constitute almost 55% of the variance.Third,fourth and fifth components has 42% of the data spread.The last component has less than 5% of the variance.Hence we can drop the fifth component

```python
pca=PCA(n_components=5)
X_new=pca.fit_transform(X)
X_new
```

```python
pca.get_covariance()
```

```python
explained_variance=pca.explained_variance_ratio_
explained_variance
```

```python
with plt.style.context('dark_background'):
    plt.figure(figsize=(6, 4))

    plt.bar(range(5), explained_variance, alpha=0.5, align='center',
            label='individual explained variance')
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc='best')
    plt.tight_layout()
```
